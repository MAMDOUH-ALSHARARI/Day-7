## Q#7: Combine the arrays z and w in horizontal way then save it in a new variable "newArray".

```python
# first solution
# to horizontally stack two arrays, they must be at same number of rows.
# I used reshape function for `z` array to be the same number of rows to `w` array
newArray = np.hstack((z.reshape(3, 1), w))
newArray
```
